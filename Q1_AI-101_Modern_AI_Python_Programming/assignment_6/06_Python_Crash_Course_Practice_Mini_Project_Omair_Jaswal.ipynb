## üî∑ Project 1: Smart Grocery Store Assistant

### üìñ Scenario:
A grocery store wants a Python assistant to manage product inventory and generate purchase receipts.

### ‚úÖ Requirements:
* Create a product_catalog dictionary: item name as key, and (price, stock) as value.
* Provide options to:
 * View available products
 * Add/update product stock or price
 * Simulate customer purchase (check stock, deduct quantity, calculate bill)
 * Generate bill with itemized prices and total
 * Print alerts for stock < 5

### üìö Helping Tips:
* Use input() to take user actions
* Use while True: loop with an exit option
* Use functions for display_products(), purchase_item(), etc.

````python
# Initialise variables
product_catalog = {}

# Fill products
product_catalog = {
  "apple": (125, 90),
  "banana": (205, 80),
  # "peach": (70, 40),
  # "guava": (75, 30),
  # "watermelon": (196, 20),
  # "cherries": (449, 35)
}


def display_products():
  print("\nProducts Catalog:")
  for product, (price, stock) in product_catalog.items():
    alert = " Low stock!" if stock < 5 else ""
    print(f"\t{product.capitalize()} - Rs.{price}, In stock: {stock}{alert}")
  print()


def add_update_product():
  product = input("\nProduct name: ").strip().lower()
  price = int(input("Product price (Rs): "))
  stock = int(input("Stock quantity: "))

  product_catalog[product] = (price, stock)
  print(f"\nCatalog updated!")
  display_products()


def purchase_items():
  cart = {}
  total = 0

  # Customer purchase
  while True:
    item = input("Enter Product Name to add item (or 'done' to finish order): ").strip().lower()

    # Check input
    if item == "done":
      break
    if item not in product_catalog:
      print("\tNot in catalog.")
      continue

    quantity = int(input(f"Quantity of {item.capitalize()}: "))

    # Get product price and stock from catalog
    price, stock = product_catalog[item]

    # Stock check
    if quantity > stock:
      print(f"\tOnly {stock} in stock.")
    else:
      # Add item to cart
      cart[item] = cart.get(item, 0) + quantity
      # Update product stock in catalog
      product_catalog[item] = (price, stock - quantity)
      # Give feedback
      print(f"\tAdded {quantity} x {item.capitalize()} to cart.")

  # Generate bill if items added in cart
  if cart:
    print("\n--- Purchase Receipt ---")

    i = 1
    # Cycle through entire cart
    for item, quantity in cart.items():
      # Get product price, ignore catalog stock
      price, _ = product_catalog[item]
      # Calculate subtotal
      line = price * quantity
      # Calculate total
      total += line
      # Give feedback
      print(f"\t{i}. {item.capitalize()} x {quantity} @ Rs.{price} = Rs.{line}")
      #
      i = i + 1

    print(f"\tTOTAL: Rs.{total}")

  print()


def main():
  print("=== Smart Grocery Store Assistant ===")

  while True:
    print("\n1. View Products\n2. Add/Update Product\n3. Purchase Items\n4. Exit")

    choice = input("Enter your choice: ").strip()
    if choice == "1":
      display_products()
    elif choice == "2":
      add_update_product()
    elif choice == "3":
      purchase_items()
    elif choice == "4":
      print("\nGoodbye!\n")
      break


if __name__ == "__main__":
  main()
````

## üî∑ Project 2: Library Book Management System

###üìñ Scenario:
Build a simple library system where users can borrow, return, or view books.

### ‚úÖ Requirements:
* Store books in a dictionary:
```
books = {
    "1984": {"author": "Orwell", "available": True},
    "Dracula": {"author": "Stoker", "available": True}
}
```
* Menu Options:
 * View all books
 * Borrow a book (check if available)
 * Return a book
 * Add a new book
 * View borrowed books

### üìö Helping Tips:
* Use .lower() for case-insensitive match
* Track borrowed books using a list or dictionary

````python
# Initialise variables
books = {}
borrowed = []


# Add books
books = {
  "1984": {
      "author": "George Orwell",
      "available": True
  },
  "Dracula": {
      "author": "Bram Stoker",
      "available": True
  },
  # "To Kill a Mockingbird": {
  #     "author": "Harper Lee",
  #     "available": True
  # },
  # "The Great Gatsby": {
  #     "author": "F. Scott Fitzgerald",
  #     "available": True
  # },
  # "The Catcher in the Rye": {
  #     "author": "J.D. Salinger",
  #     "available": True
  # }
}


def view_books():
  print("\nLibrary Catalog:")
  for title, info in books.items():
    status = "Available" if info["available"] else "Checked out"
    print(f"\t'{title}' by {info['author']}, {status}")
  print()


def add_book():
  title = input("\nBook title: ").strip()
  author = input("Author name: ").strip()

  if title in books:
    print("\tBook already in catalog.")
  else:
    books[title] = {"author": author, "available": True}
    print(f"\nCatalog update!")
    view_books()


def view_borrowed():
  print("\nBorrowed Books:")
  if borrowed:
    for title in borrowed:
      print(f"\t{title}")
  else:
    print("\tNo books currently borrowed.")
  print()


def find_match(title, catalog):
  found = None
  for book in catalog:
    if book.lower() == title.lower():
      return book


def borrow_book():
  title = input("\nEnter book title to borrow: ").strip()

  # Search book
  match = find_match(title, books)

  #
  if match is None:
    print("\tBook not found.")
  elif not books[match]["available"]:
    print("\tAlready borrowed.")
  else:
    books[match]["available"] = False
    borrowed.append(match)
    print(f"\tBook '{match}' borrowed.")


def return_book():
  title = input("\nEnter book title to return: ").strip()

  # Search book
  match = find_match(title, borrowed)

  if match is None:
    print("\tBook not found.")
  else:
    books[match]["available"] = True
    borrowed.remove(match)
    print(f"\tBook '{match}' returned.")


def main():
  print("=== Library Management System ===")

  while True:
    print("\n1. View Books\n2. View Borrowed\n3. Borrow Book\n4. Return Book\n5. Add Book\n6. Exit")

    choice = input("Enter your choice: ").strip()
    if choice == "1":
      view_books()
    elif choice == "2":
      view_borrowed()
    elif choice == "3":
      borrow_book()
    elif choice == "4":
      return_book()
    elif choice == "5":
      add_book()
    elif choice == "6":
      print("\nGoodbye!\n")
      break
    else:
      print("\nTry again!\n")
 `

if __name__ == "__main__":
  main()
````

## üî∑ Project 3: Student Report Card Generator

### üìñ Scenario:
Create a system that accepts marks for 5 subjects and generates a report card.

### ‚úÖ Requirements:
* Input: student name, roll number, and marks for 5 subjects
* Store data like:
```
students = {
    "Ali": {
        "roll": 101,
        "marks": [80, 90, 85, 88, 92]
    }
}
```
* Calculate total, average, and grade (A, B, C, D, F)
* Allow multiple student entries

### üìö Helping Tips:
* Use sum() and len() to calculate average
* Use if-else ladder for grading

````python
# Initialise variables
subjects = ["Maths", "Physics", "Chemistry", "Urdu", "English"]
students = {
  "Ali": {
    "roll": 101,
    "marks": [95, 90, 92, 93, 88]
  },
  "Zulfiqar": {
    "roll": 105,
    "marks": [50, 55, 58, 60, 52]
  },
  # "Hammaad": {
  #   "roll": 103,
  #   "marks": [70, 75, 78, 72, 80]
  # },
  "Moazzam": {
    "roll": 104,
    "marks": [60, 62, 68, 70, 69]
  },
    "Fakhar": {
    "roll": 102,
    "marks": [80, 85, 88, 82, 85]
  }
}

def view_records():
  print("\nStudent Records:")
  print(f"Name\t  | Roll | Marks")
  for name, data in students.items():
    marks_str = ", ".join(str(mark) for mark in data['marks'])
    print(f"{name:<10}| {data['roll']:<4} | {marks_str:<24}")
  print()


def add_student():
  name = input("\nStudent name: ").strip()
  roll = input("Roll number: ").strip()
  marks = []

  for i in range(0, 5):
    score = int(input(f" Marks for {subjects[i]} subject: "))
    marks.append(score)

  students[name] = {"roll": roll, "marks": marks}
  print(f"\tAdded {name} (Roll {roll}).\n")


def calculate_grade(avg):
  if avg >= 90:
    return "A"
  elif avg >= 80:
    return "B"
  elif avg >= 70:
    return "C"
  elif avg >= 60:
    return "D"
  else:
    return "F"


def generate_report():
  print("\n=== Report Cards ===")

  if not students:
    print("No student data available.\n")
    return

  for name, info in students.items():
    marks = info["marks"]
    total = sum(marks)
    average = total / len(marks)
    grade = calculate_grade(average)

    print(f"""
Name   : {name}
Roll   : {info["roll"]}
Marks  : {marks}
Total  : {total}
Average: {average}
Grade  : {grade}
------------------------""")
  print()


def main():
  print("=== Report Card Generator ===")

  while True:
    print("1. View Records\n2. Add Student\n3. Generate Reports\n4. Exit")

    choice = input("Enter your choice: ").strip()
    if choice == "1":
      view_records()
    elif choice == "2":
      add_student()
    elif choice == "3":
      generate_report()
    elif choice == "4":
      print("\nGoodbye!\n")
      break
    else:
      print("\nTry again!\n")


if __name__ == "__main__":
  main()
````

## üéÆ Bonus Project: Mini Quiz Game

### üìñ Scenario:
Build a fun Python quiz game that asks 5 questions to the user and scores them.

### ‚úÖ Requirements:
* Store questions in a dictionary or list
* Ask each question using input()
* Count the number of correct answers
* Show final score and feedback

### üìö Helping Tips:
* Use a list of tuples or dictionaries to store question-answer pairs
* Use a for loop to iterate through questions

````python
import random

questions = [
    {
        "q": "What is the capital of France?",
        "a": "paris"
    },
    {
        "q": "2 + 2 equals?",
        "a": "4"
    },
    {
        "q": "Python is a ____ language?",
        "a": "programming"
    },
    {
        "q": "Color of the sky on a clear day?",
        "a": "blue"
    },
    {
        "q": "How many continents are there?",
        "a": "7"
    },
    {
        "q": "What is the largest planet in our Solar System?",
        "a": "jupiter"
    },
    {
        "q": "What do bees collect and transform into honey?",
        "a": "nectar"
    },
    {
        "q": "Which instrument has 88 keys and black-and-white pairs?",
        "a": "piano"
    },
    {
        "q": "In computing, what does ‚ÄúCPU‚Äù stand for?",
        "a": "central processing unit"
    },
    {
        "q": "What‚Äôs the name of the toy that builds with interlocking plastic bricks?",
        "a": "lego"
    }
]


def quiz():
  score = 0

  # Randomly pick 5 distinct questions
  subset = random.sample(questions, 5)

  for i, qa in enumerate(subset, start=1):
    print(f"\nQ{i}: {qa['q']} ")
    answer = input("Your Answer: ").strip().lower()
    if answer == qa["a"]:
      print("Correct!")
      score += 1
    else:
      print(f"Wrong! (Ans: {qa['a'].capitalize()})")

  print(f"\nFinal Score: {score}/{len(subset)}")

  feedback = "Excellent job!" if score > 4 else "Keep practicing!"
  print(feedback)


def main():
  print("=== Mini Quiz Game ===")
  quiz()


if __name__ == "__main__":
  main()
````
